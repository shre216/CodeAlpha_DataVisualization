```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Global Superstore.csv")
df.head()
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt

df.groupby('Category')['Sales'].sum().plot(kind='bar', title='Sales by Category')
plt.show()

df.groupby('Region')['Sales'].sum().plot(kind='pie', autopct='%1.1f%%')
plt.title("Sales by Region")
plt.show()

import matplotlib.pyplot as plt

# Convert to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')

# Remove missing dates
df = df.dropna(subset=['Order Date'])

# Create Year column
df['Year'] = df['Order Date'].dt.year

# Group and plot
df.groupby('Year')['Sales'].sum().plot(kind='line', marker='o')

plt.title("Sales Over Years")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.show()

import seaborn as sns

sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap")
plt.show()

top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

top_products.plot(kind='bar', title='Top 10 Products by Sales')
import matplotlib.pyplot as plt
plt.xticks(rotation=45)
plt.show()
```
